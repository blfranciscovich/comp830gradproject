## 3b 
### Constructors
In Java and Python, constructors have the same name as the class and are declared public. Constructors allow programmers to get an instance of the class or for example, for Apple to get an instance of their cellphone blueprint.

JAVA They are declared without a return type. So any function that is named the same as the class and has no return type is a constructor. Our constructor will take two parameters the numerator and the denominator.

PYTHON Every class has a factory, called by the same name as the class, for making new instances. If the class has an initializer method, this method is used to get the attributes (i.e. the state) of the new object properly set up.

### Setters and Getters


#### Python Code
To run this code, make sure you are in Level 2 Python

```python
grade = 95

#Divide grade by 10 and get disregard the remainer
letterGrade = grade // 10

switch = {
    10: 'Perfect Score!',
    9: 'A',
    8: 'B',
    7: 'C',
    6: 'D',
}

# Get the letter grade then state 'F' as the default
print (switch.get(letterGrade, 'F'))
```

#### Python Code
To run this code, make sure you are in Level 2 Python

```python
#Let's write a for loop
for i in range(10):
    print(i)
#So we can see the different loops, let's put a space between them
print('\n')

#Let's use more of the range functions
for i in range(0, 10):
    print(i)
print('\n')

#Let's use all of the range functions
for i in range(0, 10, 1):
    print(i)
print('\n')

#Let's write a while loop
i = 10
while (i > 0):
    i -= 1
    print(i)
```
